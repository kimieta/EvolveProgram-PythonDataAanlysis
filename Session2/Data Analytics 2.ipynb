#### Scatter Graph

```python
import numpy as np
from numpy import random

# np.random.normal(0, 2, 24) generates 24 random numbers from a normal distribution with a mean of 0 and a standard deviation of 2.
# Adding 4 to this result shifts the mean of the distribution from 0 to 4. This means that the generated numbers will now have a mean of 4, but they will still have a standard deviation of 2.
# The result is an array x2 of length 24 with values drawn from a normal distribution centered around 4.

x2 = 4 + np.random.normal(0, 2, 24)
y2 = 4 + np.random.normal(0, 2, len(x2))

plt.scatter(x2,y2)
```

#### Fill Graph

```python
x4 = np.linspace(0.1, 9.9, 20) # Create 20 evenly spaced points between 0.1 and 9.9
y4 = 3.0 * x4 # Multiply each of those points by 3 to create a linear relationship

ci = 1.96 * np.std(y4) / np.sqrt(len(x4)) # Calculate the confidence interval

fig, ax = plt.subplots() # Create a plot
ax.plot(x4, y4) # Plot the original data points
ax.fill_between(x4, (y4-ci), (y4+ci), color='b', alpha=.1) # Fill the area between (y1 - ci) and (y1 + ci)
```

```python
x = np.linspace(0, 10, 11)  # Create 11 evenly spaced points between 0 and 10
y = [3.9, 4.4, 10.8, 10.3, 11.2, 13.1, 14.1, 9.9, 13.9, 15.1, 12.5]  # Predefined y-values

# fitting a linear curve
a, b = np.polyfit(x, y, deg=1)  # Fit a linear polynomial (degree 1) to the x and y data
y_est = a * x + b  # Calculate the estimated y-values based on the fitted line

# calculating error (confidence band)
y_err = np.std(y) * np.sqrt(1/len(x) + (x - x.mean())**2 / np.sum((x - x.mean())**2)) 
# np.std(y) calculates the standard deviation of the y-values.
# np.sqrt(1/len(x) + (x - x.mean())**2 / np.sum((x - x.mean())**2)) calculates the error for each x value to create the confidence band.

fig, ax = plt.subplots()  # Create a plot
ax.plot(x, y_est, '-')  # Plot the linear regression line
ax.fill_between(x, y_est - y_err, y_est + y_err, alpha=0.2)  # Fill the area between (y_est - y_err) and (y_est + y_err) with 20% transparency
ax.plot(x, y, 'o', color='tab:brown')  # Plot the original data points as brown dots
```

#### Heatmap

```python
# Create a 10x10 matrix of random values
data = np.random.rand(10, 10)

# Create the heatmap
plt.imshow(data, cmap='viridis', interpolation='nearest')
# interpolation='nearest' ensures that each cell is colored distinctly.

# Add a color bar to the right of the heatmap
plt.colorbar()

# Show the heatmap
plt.show()
```

```python
# Days of the week and time slots
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
time_slots = ["Breakfast", "Lunch", "Dinner"]

# Random ratings data (7 days x 3 time slots)
ratings = np.array([
    [4, 3, 5],  # Monday
    [2, 4, 4],  # Tuesday
    [3, 2, 5],  # Wednesday
    [4, 3, 4],  # Thursday
    [5, 4, 3],  # Friday
    [3, 5, 4],  # Saturday
    [4, 4, 5],  # Sunday
])

# Create the heatmap
fig, ax = plt.subplots(figsize=(8, 6))
cax = ax.matshow(ratings, cmap='coolwarm')

# Add color bar
fig.colorbar(cax)

# Set axis labels
ax.set_xticks(np.arange(len(time_slots)))
ax.set_yticks(np.arange(len(days)))
ax.set_xticklabels(time_slots)
ax.set_yticklabels(days)

# Rotate the tick labels and set their alignment
plt.xticks(rotation=45, ha='left')

# Annotate each cell with the numerical rating value
for i in range(len(days)):
    for j in range(len(time_slots)):
        ax.text(j, i, ratings[i, j], ha='center', va='center', color='black')

# Add labels and title
plt.xlabel('Time Slot')
plt.ylabel('Day of the Week')
plt.title('Restaurant Ratings Heatmap')

# Show the heatmap
plt.show()
```

#### Boxplot

```python
# Sample data
data = np.random.normal(0, 1, 100)

# Create a boxplot
plt.figure()
plt.boxplot(data)
plt.title('Boxplot Example')
plt.ylabel('Values')
plt.show()
```

#### Error bar

```python
# Sample data
x = np.linspace(0, 10, 10)
y = np.sin(x)
yerr = 0.2  # Error

# Create an error bar plot
plt.figure()
plt.errorbar(x, y, yerr=yerr, fmt='o', ecolor='r', capsize=5)
plt.title('Error Bar Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

#### Event Plot

```python
# Sample data
data = [np.random.random(10) for _ in range(3)]

# Create an event plot
plt.figure()
plt.eventplot(data, orientation='horizontal', colors='b')
plt.title('Event Plot Example')
plt.xlabel('Event Time')
plt.show()
```

#### Violin Plot

```python
# Sample data
data = [np.random.normal(0, std, 100) for std in range(1, 5)]

# Create a violin plot
plt.figure()
plt.violinplot(data, showmeans=False, showmedians=True)
plt.title('Violin Plot Example')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

#### Pie Chart

```python
# Sample data
labels = 'A', 'B', 'C', 'D'
sizes = [15, 30, 45, 10]

# Create a pie chart
plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Example')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

#### Cumulative Distribution 

```python
# Sample data
data = np.random.randn(1000)

# Create a cumulative distribution plot
plt.figure()
plt.hist(data, bins=30, cumulative=True, color='blue', edgecolor='black', alpha=0.7)
plt.title('Cumulative Distribution Plot Example')
plt.xlabel('Values')
plt.ylabel('Cumulative Frequency')
plt.show()
```
